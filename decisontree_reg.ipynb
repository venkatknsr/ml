```python
import pandas as pd
import numpy as np
df=pd.read_csv(r'/content/ML Dataset.csv')
df
new_df=df.drop(['Area','BHK'],axis=1)
new_df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
label=le.fit_transform(new_df['Furnished'])
label
```

```python
new_df.drop('Furnished',axis=1,inplace=True)
new_df['Furnished']=label
new_df
```

```python
le1=LabelEncoder()
label2=le1.fit_transform(new_df['Property Type'])
label2
```

```python
new_df.drop('Property Type',axis=1,inplace=True)
new_df['Property Type']=label2
new_df
```

```python
X=new_df[['Sq.ft','Furnished','Property Type']]
y=new_df['Rent']
```

```python
from sklearn.tree import DecisionTreeRegressor
re=DecisionTreeRegressor(random_state=0)
re.fit(X,y)
```

```python
print(re.predict([[2000,0,0]]))
```

```python
from matplotlib import pyplot as plt
plt.plot(X,re.predict(X),color='red')
```
