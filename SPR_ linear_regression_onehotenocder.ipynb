```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Ml Dataset.csv')
data
```

```python
#using one hot encoder
one_hot_encoded_data = pd.get_dummies(data, columns = ['Area','Furnished','Property Type'])
df1=one_hot_encoded_data.astype(int)
df1
```

```python
df1.drop(['Property Type_Independent'],axis=1,inplace=True)
df1
```

```python
#for getting the column index
def get_column_index(df, column_name):
    if column_name in df.columns:
        return list(df.columns).index(column_name)
    else:
        return None 
```

```python
#creating a list to give it to the model for prediction
print(df1.shape[1])
lisa=[0]*15

```

```python
lisa=[0]*15
#taking sft input from user
sft=int(input())
s=input()
#taking area input from user
s1="Area_"+s
index_area=get_column_index(df1,s1)
#taking furniture_type input from user
s=input()
s2="Furnished_"+s
index_furniture=get_column_index(df1,s2)
print(index_furniture)
#taking property_type input from user
s=input()
s2="Property Type_"+s
index_property=get_column_index(df1,s2)
print(index_property)
```

```python
#assigning thoses values to the list at particular indexes
lisa[0]=sft
lisa[index_area]=1
lisa[index_furniture]=1
print("flag is 0 for independent else flag is 1 for apartment")
flag=int(input())
lisa[index_property]=flag
print(lisa)
```

```python
cols = list(df1.columns)

# Remove the column name from this list
cols.remove('Rent')

# Append it back to the end
cols.append('Rent')
df1 = df1[cols]
df1
```

```python
#intializing independent and dependent features
X=df1.iloc[:,:-1]
print(len(lisa))
y=df1.iloc[:,-1]
print(X.shape)
```

```python
#using linear regression model
from sklearn import linear_model
reg=linear_model.LinearRegression()
reg.fit(X,y)
```

```python
#predicting the rent price for given input by the user
reg.predict([lisa])
```
